# Win/Loss ratio's correlation

For each sport, we will find the win/loss ratio's correlation with the population of the city it is in. Win/Loss ratio refers to the number of wins over the number of wins plus the number of losses. We will calculate the correlation with pearsonr using the population data obtained from the wikipedia_data.html file and the win/loss ratio for a given sport.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import re

mlb_df=pd.read_csv("mlb.csv")
nhl_df=pd.read_csv("nhl.csv")
nba_df=pd.read_csv("nba.csv")
nfl_df=pd.read_csv("nfl.csv")
cities=pd.read_html("wikipedia_data.html")[1]
cities=cities.iloc[:-1,[0,3,5,6,7,8]]
```

##  NHL
The win/loss ratio's correlation with the population of the city it is in for the NHL using 2018 data. 

```python
def nhl_helper():
    cities_copy = cities[['Metropolitan area', 'Population (2016 est.)[8]', 'NHL']].copy()
    cities_copy['NHL'] = cities_copy['NHL'].replace('\[.*\]|—', '', regex=True)
    cities_copy = cities_copy[cities_copy['NHL'].astype(bool)]
    
    nyc_teams = cities_copy.iloc[0]['NHL']
    nyc_teams = re.findall('[A-Z][a-z]+', nyc_teams)
    la_teams = cities_copy.loc[1]['NHL']
    la_teams = re.findall('[A-Z][a-z]+', la_teams)

    for team in nyc_teams:
        t = cities_copy.iloc[0].copy()
        t['NHL'] = team
        cities_copy = cities_copy.append(t, ignore_index=True)
    for team in la_teams:
        t = cities_copy.iloc[1].copy()
        t['NHL'] = team
        cities_copy = cities_copy.append(t, ignore_index=True)

    cities_copy = cities_copy.drop([0,1])
    
    nhl_df_copy = nhl_df[nhl_df['year'] == 2018].copy()
    nhl_df_copy[['W', 'L']] = nhl_df_copy[['W', 'L']].replace('[a-zA-Z]+', np.nan, regex=True)
    nhl_df_copy.dropna(inplace=True)
    nhl_df_copy.index = range(len(nhl_df_copy.index))
    nhl_df_copy[['W', 'L']] = nhl_df_copy[['W', 'L']].astype(int)
    nhl_df_copy['win_loss_ratio'] = nhl_df_copy['W'] / (nhl_df_copy['W'] + nhl_df_copy['L'])
    
    
    def w_l_ratio(x):
        teams = list(nhl_df_copy['team'])
        for i in range(len(teams)):
            if x in teams[i]:
                return nhl_df_copy.iloc[i]['win_loss_ratio']
    
    cities_copy['win_loss_ratio'] = cities_copy['NHL'].apply(lambda x : w_l_ratio(x))
    cities_copy['Population (2016 est.)[8]'] = cities_copy['Population (2016 est.)[8]'].astype(int)
    cities_copy = cities_copy.groupby('Metropolitan area').agg({'Population (2016 est.)[8]': np.mean,'win_loss_ratio': np.mean})
    cities_copy.rename(columns={'Population (2016 est.)[8]': 'Population'}, inplace=True)
    
    return cities_copy

def nhl_correlation():     
    cities_copy = nhl_helper()
    
    population_by_region = cities_copy['Population'] 
    win_loss_by_region = cities_copy['win_loss_ratio']

    return stats.pearsonr(population_by_region, win_loss_by_region)[0]

nhl_correlation()
```

## Hypothesis
Given that an area has two sports teams in different sports, we will explore if those teams will perform the same within their respective sports. So we wil explore this idea with a series of paired t-tests between all pairs of sports.

```python
def sports_team_performance():
    mlb_df = mlb_helper()
    mlb_df.name = 'MLB'
    
    nhl_df = nhl_helper()
    nhl_df.name = 'NHL'
    
    nba_df = nba_helper()
    nba_df.name = 'NBA'
    
    nfl_df = nfl_helper()
    nfl_df.name = 'NFL'
    
    sports = ['NFL', 'NBA', 'NHL', 'MLB']
    p_values = pd.DataFrame({k:np.nan for k in sports}, index=sports)

    sports_df = [nfl_df, nba_df, nhl_df, mlb_df]
    for row in sports_df:
        for col in sports_df:
            if row.name == col.name:
                p_values.loc[row.name, col.name] = np.nan
            else:
                df = pd.merge(row, col, how='inner', left_index=True, right_index=True)
                p_val = stats.ttest_rel(df['win_loss_ratio_x'], df['win_loss_ratio_y'])
                p_values.loc[row.name][col.name] = p_val[1]
    
    return p_values

sports_team_performance()
```
